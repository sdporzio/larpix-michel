```python
%%time
### EVENT DATAFRAME
base_events = pd.DataFrame()
# Add all standard columns
for n in tqdm(h_events.dtype.names):
    base_events[n] = h_events[n]
# Add custom columns: Length of longest track
maxLength = []
for e in tqdm(h_events):
    if e['ntracks']>0: maxLength.append(np.max(h_tracks[e['track_ref']]['length']))
    else: maxLength.append(0)
base_events['maxLength'] = maxLength

    
### TRACK DATAFRAME
base_tracks = pd.DataFrame()
# Add all standard columns
# >1D arrays need special treatment (.tolist)
for n in tqdm(h_tracks.dtype.names):
    if n=='residual' or n=='start' or n=='end': base_tracks[n] = h_tracks[n].tolist()
    else: base_tracks[n] = h_tracks[n]
# Add custom columns: Useful values from event array
evid,ntracks,nhits,longestTrackInEv = [], [], [], []
for r in tqdm(h_tracks['event_ref']):
    ev = h_events[r]
    evid.append(ev['evid'][0])
    nhits.append(ev['nhit'][0])
    ntracks.append(ev['ntracks'][0])
base_tracks['evid'] = evid
base_tracks['nhits'] = nhits
base_tracks['ntracks'] = ntracks
```
